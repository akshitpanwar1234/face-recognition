```
import face_recognition
image1=face_recognition.load_image_file("/content/drive/My Drive/Ch06/person_1.jpg")
image2=face_recognition.load_image_file("/content/drive/My Drive/Ch06/person_2.jpg")
image3=face_recognition.load_image_file("/content/drive/My Drive/Ch06/person_3.jpg")
```

```
person1=face_recognition.face_encodings(image1)
person2=face_recognition.face_encodings(image2)
person3=face_recognition.face_encodings(image3)
```

```
knownface=[person1,person2,person3]
```

```
notknown = face_recognition.load_image_file("/content/drive/My Drive/Ch06/unknown_8.jpg")
notknownimage = face_recognition.face_encodings(notknown)
for notknownimages in notknownimage:
  i=1
  if face_recognition.compare_faces(knownface,notknownimages,tolerance=0.6):
    print(f"person {i} found")
    i+=1
  else:
    print("perrson {i} not found")
    i+=1
```
